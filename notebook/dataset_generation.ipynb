```python
import random
import numpy as np
from igsm_gym.generation import ProblemGenerator
import json
```

```python
train_num = 100000
test_num = 1000
```

```python
# Initialize an empty list to store the data
data = []

seed = 0
data_idx = 1
while data_idx <= train_num:
    if data_idx % 100 == 0:
        print("data_idx:", data_idx)
    seed += 1
    random.seed(seed)
    np.random.seed(seed)
    pg = ProblemGenerator(debug=True)
    while not pg.draw_question():
        continue
    
    question = pg.generate_question()
    answer = pg.generate_answer()
    
    ## Save data
    entry = {
        "index": data_idx,
        "query": question,
        "response": answer,
        "seed": seed
    }
    data.append(entry)
    
    data_idx += 1
    
# Save the data to a JSON file
output_file = "iGSM_train.json"
with open(output_file, "w") as f:
    json.dump(data, f, indent=4)  # `indent=4` makes the JSON file more readable
```

```python
# Initialize an empty list to store the data
data = []

# Testing

seed = 100000
data_idx = 1
while data_idx <= test_num:
    if data_idx % 100 == 0:
        print("data_idx:", data_idx)
    seed += 1
    random.seed(seed)
    np.random.seed(seed)
    pg = ProblemGenerator(debug=True)
    while not pg.draw_question():
        continue
    
    question = pg.generate_question()
    answer = pg.generate_answer()
    
    answer = pg.Gd.topo[-1].value
    
    ## Save data
    entry = {
        "index": data_idx,
        "question": question,
        "answer": answer,
        "seed": seed,
        
    }
    data.append(entry)
    
    data_idx += 1
    
# Save the data to a JSONL file
output_file = "iGSM_test_1K.jsonl"
with open(output_file, "w") as f:
    for entry in data:
        f.write(json.dumps(entry) + "\n")  # Convert each entry to a JSON string and write it as a new line
```
